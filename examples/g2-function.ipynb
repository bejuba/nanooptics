# reading and calculating the g2-function from TCSPC data

```python
import matplotlib.pyplot as plt
from nanooptics import read_tcspc, cycorr
```

```python
header, data = read_tcspc.read_picoquant('./sample_data.pt2')
result = cycorr.corr(channel=data.channel, timestamp=data.timestamp, cutofftime=500e-9, resolution=800e-12)
```

```python
print(header)
plt.errorbar(result.tau, result.g2, result.yerr, fmt='.')
```
